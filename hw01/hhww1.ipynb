## 3. Latex formula and plot
---
#### *Requirement: Create a series x of range 0 to 1000 and a series y which is any non-linear function of x (Hint: if you cant' think of a fun formula, stick with the quadratic function). Use the plotting package of your choice to graph the function. Add the latex formula of the function you plotted in the text below.**

```python
# Import Python Libraries
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
import seaborn as sns

import matplotlib.pyplot as plt

# Create a series of x of range 0 to 1000
x = pd.Series(range(0,1000+1))

# Create a series y which is any non-linear function of x
y = x**2 + x

# Plotting the function
plt.figure(figsize=(12, 10))
plt.plot(x, y, label= "y = x^2 + x")
plt.title("Plot of the function y = x^2 + x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
