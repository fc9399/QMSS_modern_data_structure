#### 3. The trial

Let's take a look at the information on the `trial`.

a) What are the average and maximum numbers of male and female accusers?  
b) Count the number of `sentence`s by sentence type. List them in a table (in descending order), excluding missing values. Rename the column headings to something sensible.  
c) Do the number of accusers matter for the `verdict`? Compare the average number of accusers by the type of verdict. Again make sure the table is sorted and the headings make sense. 

```python
%%sql
# Check the table first
SELECT * FROM trial LIMIT 3;
```

a) What are the average and maximum numbers of male and female accusers?

```python
%%sql
SELECT
    AVG(male_accusers) AS avg_male_accusers,
    MAX(male_accusers) AS max_male_accusers,
    AVG(female_accusers) AS avg_female_accusers,
    MAX(female_accusers) AS max_female_accusers
FROM trial;
```

b) Count the number of `sentence`s by sentence type. List them in a table (in descending order), excluding missing values. Rename the column headings to something sensible. 

```python
%%sql
SELECT 
    sentence AS sentence_type,
    COUNT(*) AS sentence_count
FROM trial
WHERE sentence IS NOT NULL
GROUP BY sentence_type
ORDER BY sentence_count DESC;
```

c) Do the number of accusers matter for the `verdict`? Compare the average number of accusers by the type of verdict. Again make sure the table is sorted and the headings make sense.

```python
%%sql
SELECT
    verdict AS verdict_type,
    AVG(male_accusers + female_accusers) AS avg_accusers
FROM trial
WHERE verdict IS NOT NULL # note: Remember I cannot use the"verdict_type" renamed in SELECT clause because WHERE clause is executed before SELECT clause
GROUP BY verdict_type
ORDER BY avg_accusers DESC;
```

#### 4. Tortured Truth

a) Left join the `trial` and `confession` tables. For what share of trials does the database record confessions? Create a results table with the number of all trials, the number of confessions, and the share of trials with confessions recorded. 

```python
%%sql
#Check trial table to search for similarity
SELECT * FROM trial
LIMIT 1;
```

```python
%%sql
#Check confession table to search for similarity
SELECT * FROM confession
LIMIT 1;
```

Both tables have trialref, so this will be the point of joinning two tables

```python
%%sql
SELECT 
    COUNT(DISTINCT trial.trialref) AS total_trials,
    COUNT(DISTINCT confession.trialref) AS total_confessions,
    (COUNT(DISTINCT confession.trialref) * 100.0 / COUNT(DISTINCT trial.trialref)) AS share_with_confessions_percent
FROM trial
LEFT JOIN confession ON trial.trialref = confession.trialref;
```


b) Only a small number of trials have records of torture. Is there a higher share of confessions among trials with records of torture than trials without such record? _Hint:_ You will need to merge on the `confession` table. 

```python
%%sql
# Check torture table
SELECT * FROM torture 
LIMIT 3;
```

My method: The torture table is joined to the trial table via trialref. The condition checks if the trialref in the torture table exists (i.e., is not null). If it exists, it indicates that there is a record of torture associated with this trial.

```python
%%sql
SELECT 
    # if the final table has torture table's trial ref (not null), then there was a torture in that trial.
    CASE WHEN torture.trialref IS NOT NULL THEN 'With Torture' ELSE 'Without Torture' END AS torture_status,
    COUNT(DISTINCT trial.trialref) AS total_trials,
    COUNT(DISTINCT confession.trialref) AS total_confessions,
    (COUNT(DISTINCT confession.trialref) * 100 / COUNT(DISTINCT trial.trialref)) AS share_with_confessions_percent
FROM trial
LEFT JOIN torture ON trial.trialref = torture.trialref
LEFT JOIN confession ON trial.trialref = confession.trialref
GROUP BY torture_status;
```

Conclusion:

Based the result table above, there is a significantly higher share of confessions among trials with records of torture than trials without such record.
